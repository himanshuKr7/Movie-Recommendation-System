```python
def convert(obj):
        L=[]
        counter=0
        for i in ast.literal_eval(obj):
            if counter<3:
                L.append(i['name'])
                counter+=1
            else:
                break
        return L
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
def getdirector(obj):
        L=[]
        for i in ast.literal_eval(obj): #here using this module string will be converted to list
            if i['job']=='Director':
                L.append(i['name'])
                break
        return L
```

```python
movies['crew']=movies['crew'].apply(getdirector)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) #converting the content of overview to list    
```

```python
movies.head()
```

```python
#to remove all the space between all the strings
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) #taking out of all the elements in list
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower()) #converting all the content of tags in smaller case
```

```python
new_df.head()
```

```python
import nltk  #steming of the word is done
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer( )
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# text vectorization

```python
#doing text vectorization by importing a libraray countvectorizer from sklearn
#STOP WORDS REMOVAL

#We will not consider the "STOP WORDS" for the vectorization process so they will be removed from the dataframe using the CountVectorizer class.

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#transforming the tags coloumn to array
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
 sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]   #getting te list of 5  similar movies enumerate is used for creating list according to index basis
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
                         
    for i in movies_list:
                         print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
